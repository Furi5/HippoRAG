```python
from llama_index.embeddings.ollama import OllamaEmbedding
embedding_model = OllamaEmbedding(
            model_name="nomic-embed-text:latest", 
            base_url="http://localhost:11434"
            )

```

```python
import numpy as np

raw_embeddings = embedding_model.get_text_embedding_batch(['None','' 'the quick brown fox jumps over the lazy dog'])
raw_embeddings[1]
```

```python
for i in range(len(raw_embeddings)):
    if raw_embeddings[i] is None:
        raw_embeddings[i] = np.zeros(768, dtype=np.float32)
```

```python
np.array(raw_embeddings).shape
```

```python
embedding = np.array(raw_embeddings)
embedding = (embedding.T / np.linalg.norm(embedding, axis=1)).T
```
